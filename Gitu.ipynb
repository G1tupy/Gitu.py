```python
import numpy as np
import matplotlib.pyplot as plt
# Sample student data
students = ['Jhon', 'Smith', 'Marry', 'Rose', 'Devid']
math_scores = [85, 92, 78, 88, 90]
science_scores = [76, 88, 92, 80, 78]
# Create a subplot with two subplots (1 row, 2 columns)
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
# Plot Math Scores
axs[0].bar(students, math_scores, color='r')
axs[0].set_title('Math Scores')
axs[0].set_xlabel('Students')
axs[0].set_ylabel('Score')
# Plot Science Scores
axs[1].bar(students, science_scores, color='b')
axs[1].set_title('Science Scores')
axs[1].set_xlabel('Students')
axs[1].set_ylabel('Score')
# Adjust spacing between subplots
plt.tight_layout()
# Show the subplot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample sales data (replace this with your actual data)
data = {
    'Date': pd.date_range(start='2022-01-01', periods=10, freq='M'),
    'Region_A': [100, 120, 130, 110, 150, 140, 160, 170, 180, 200],
    'Region_B': [80, 90, 100, 110, 120, 130, 140, 150, 160, 170],
    'Region_C': [50, 60, 70, 80, 90, 100, 110, 120, 130, 140],
}

df = pd.DataFrame(data)
df.set_index('Date', inplace=True)

# Create subplots
fig, axs = plt.subplots(nrows=3, ncols=1, figsize=(10, 12), sharex=True)

# Plot each region on a separate subplot
axs[0].plot(df.index, df['Region_A'], label='Region A', marker='o')
axs[0].set_ylabel('Sales')
axs[0].legend()

axs[1].plot(df.index, df['Region_B'], label='Region B', marker='o')
axs[1].set_ylabel('Sales')
axs[1].legend()

axs[2].plot(df.index, df['Region_C'], label='Region C', marker='o')
axs[2].set_xlabel('Date')
axs[2].set_ylabel('Sales')
axs[2].legend()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset from Seaborn
iris = sns.load_dataset('iris')

# Display the first few rows of the dataset
print("First few rows of the Iris dataset:")
print(iris.head())

# Summary statistics of the dataset
print("\nSummary statistics of the Iris dataset:")
print(iris.describe())

# Count of each species in the dataset
species_count = iris['species'].value_counts()
print("\nCount of each species:")
print(species_count)

# Visualization: Pairplot
sns.pairplot(iris, hue='species')
plt.title('Pairplot of Iris Dataset')
plt.show()
```

```python
import numpy as np #Import Numpy as np

# Input: Array of daily temperature readings
temperatures = np.array([32.5, 34.2, 36.8, 29.3, 31.0, 38.7, 23.1, 18.5, 22.8, 37.2,4,25,12,-4,-12])

hot_days_indices = np.where(temperatures > 35)[0] # Identify hot days (temperature > 35 degrees Celsius)
hot_days_temperatures = temperatures[hot_days_indices]

cold_days_indices = np.where(temperatures < 5)[0] # Identify cold days (temperature < 5 degrees Celsius)
cold_days_temperatures = temperatures[cold_days_indices]

print("Hot Days:") # Display the result of hot days
print("Day\tTemperature (°C)") # Display the result of temperature
for day, temp in zip(hot_days_indices + 1, hot_days_temperatures):
    print(f"{day}\t{temp:.1f}")

print("\nCold Days:")     # Display cold days and corresponding temperatures
print("Day\tTemperature (°C)")
for day, temp in zip(cold_days_indices + 1, cold_days_temperatures):
    print(f"{day}\t{temp:.1f}")

```
